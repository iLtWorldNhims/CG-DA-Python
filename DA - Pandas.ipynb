```python
data = pd.read_csv('FoodPrice_in_Turkey.csv',
                  header = 0, encoding = 'ISO-8859-1')
print('*'*20,'Nhập dữ liệu thành công','*'*20)
display(data)
data.info()
```

```python
# Đổi tên tất cả cột trong dữ liệu. Yêu cầu liệt kê đầy đủ các cột.
data.columns = ['Địa điểm','Mã SP','Tên SP','Đơn vị','Tên đơn vị','Tháng','Năm','Giá']
display(data)
```

```python
# Đổi tên 1 hoặc nhiều cột trong dữ liệu
data.rename(columns = {'UmID':'Đơn vị','UmName':'Tên đơn vị'}, inplace = True)
# Cấu trúc inplace ảnh hưởng đến việc thay đổi bộ nhớ. 
# Nếu là True => Thay đổi trên dữ liệu gốc.
# Nếu là False => Thường được gán cho 1 vùng nhớ mới. 
data
```

```python
# Thêm cột mới vào sau cùng
data['New'] = np.NaN
data
```

```python
# Chèn một cột mới vào giữa
data.insert(8,'Chiết khấu(%)',20)
data.insert(7,'Số lượng',50)
display(data)
# Cấu trúc gồm vị trí chèn, tên cột, giá trị cột. Giá trị cột có thể thêm nhiều giá trị khác nhau.
```

```python
data.rename(columns = {'New':'Tổng'}, inplace = True)
display(data)
```

```python
data['Tổng'] = data['Giá']*data['Số lượng']*(100-data['Chiết khấu(%)'])/100
display(data)
```
