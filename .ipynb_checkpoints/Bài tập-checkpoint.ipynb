```python
result = []
for number in range (2000, 3201):
    if number % 7 == 0 and number % 5 != 0:
        result.append(str(number))

print(", ".join(result))
```

```python
import random
N = int(input("Nhập một số N: "))
for i in range(N):
    Number = random.randint(0, 9999999999)
    if len(str(Number)) == 10 and int(str(Number)[0]) in [7, 8, 9]:
        print(f"{Number}: YES")
    else:
        print(f"{Number}: NO")
```
