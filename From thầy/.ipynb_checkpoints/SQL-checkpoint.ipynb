```python
#Truy xuất tên khách hàng, nước và thành phố
# mà nước là Germany hoặc USA
# sắp xếp theo thứ tự ABC tên nước
# sắp xếp theo thứ tự ABC tên thành phố
query = "SELECT customerName, country, city FROM customers \
WHERE country = 'Germany' OR country = 'USA' \
ORDER BY country, city ASC"
vidu8 = pd.read_sql_query(query, connection)
vidu8
```

## Từ khóa LIMIT

```python
#Truy xuất tên khách hàng, nước và thành phố
# từ bảng customers
# mà nước là Germany hoặc USA
# sắp xếp theo thứ tự ABC tên thành phố
# lấy 3 dữ record đầu tiên
query = "SELECT customerName, country, city FROM customers \
WHERE country = 'Germany' OR country = 'USA' \
ORDER BY city ASC \
LIMIT 3"
vidu9 = pd.read_sql_query(query, connection)
vidu9
```

```python
#Truy xuất tên khách hàng, nước, thành phố và creditLimit
# từ bảng customers
# mà nước là UK
# và có creditLimit là lớn nhất
query = "SELECT customerName, country, city, creditLimit \
FROM customers \
WHERE country = 'UK' \
ORDER BY creditLimit DESC \
LIMIT 1"
vd10 = pd.read_sql_query(query,connection)
vd10
```

## Aggregate Functions

### Hàm COUNT()

```python
#Truy xuất số lượng các thành phố của Japan từ trong bảng customers

query = "SELECT count(DISTINCT city) FROM customers WHERE country='Germany'"
vd13 = pd.read_sql_query(query,connection)
vd13
```

```python
#Truy xuất số lượng khách hàng đến từ USA

query = "select count(customerNumber) from customers where country='USA'"

vd14 = pd.read_sql_query(query,connection)
vd14
```

### Hàm SUM()

```python
#Tính tổng creditLimit của các khách hàng đến từ USA
query = "SELECT SUM(creditLimit) FROM customers WHERE country='USA'"

vd14 = pd.read_sql_query(query,connection)
vd14
```

### Hàm MIN()

```python
#Tính creditLimit nhỏ nhất của các khách hàng đến từ USA
query = "SELECT MIN(creditLimit) FROM customers WHERE country='USA'"

vd14 = pd.read_sql_query(query,connection)
vd14
```

### Hàm MAX()

```python
#Tính creditLimit nhỏ nhất của các khách hàng đến từ USA
query = "SELECT MAX(creditLimit) FROM customers WHERE country='USA'"

vd14 = pd.read_sql_query(query,connection)
vd14
```

### Hàm AVG()

```python
#Tính creditLimit nhỏ nhất của các khách hàng đến từ USA
query = "SELECT AVG(creditLimit) FROM customers WHERE country='USA'"

vd14 = pd.read_sql_query(query,connection)
vd14
```

## Từ khóa GROUP BY

```python
#Số lượng khách hàng ở từng quốc gia là bao nhiêu?
# Sắp xếp theo số lượng khách hàng giảm dần
query = "SELECT country, COUNT(country) as sl \
FROM customers \
GROUP BY country \
ORDER BY sl DESC"

vd = pd.read_sql_query(query,connection)
vd
```

```python
# Truy xuất quốc gia, thành phố và số lượng khách hàng tương ứng là bao nhiêu?
# Sắp xếp theo tên quốc gia, và số lượng khách giảm dần
query = "SELECT country, city, count(customerNumber) as sl \
FROM customers \
GROUP BY country, city \
ORDER BY country, sl DESC"

vd = pd.read_sql_query(query,connection)
vd
```

## Từ khóa HAVING

```python
# Liệt kê ra các nước có nhiều hơn 5 khách hàng từ bảng customers
```

```python
query = "SELECT country \
FROM customers \
GROUP BY country \
HAVING COUNT(country)>5"

vd = pd.read_sql_query(query,connection)
vd
```

```python
# Liệt kê ra các thành phố có creditLimit trung bình của khách hàng > 100000
query = "SELECT city \
FROM customers \
GROUP BY city \
HAVING AVG(creditLimit) > 100000"

vd = pd.read_sql_query(query,connection)
vd
```
