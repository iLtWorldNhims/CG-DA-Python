## Demo 3: Gradient Descent (làm thủ công)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print("Khai báo thành công")
```

```python
dat3 = pd.read_csv('anscombe.csv')
dat3
```

```python
plt.scatter(dat3['x'], dat3['y']);
```

```python
X = np.asarray(dat3['x'])
Y = np.asarray(dat3['y'])
```

```python
#Tạo array Y_ (giá trị tính toán)
Y_ = np.array([0])
```

```python
#Thiết lập giá trị alpha - learning rate
alpha = 0.01
n = 5000 #Số lần học
```

```python
import time as dt
```

```python
start = dt.time() #tạo mốc thời gian --> xác định thời gian tính toán
#Tạo các giá trị ban đầu của w0 và w1
w0_ = 0
w1_ = 0
#Tạo array w0 và w1
w0 = np.array([])
w1 = np.array([])
L = np.array([])
for i in range(n):
    y_ = w0_ + w1_*X
    # np.append(Y_, )
    # print(y_)
    L = np.append(L, 1/2*sum((y_-Y)**2)) #Tính hàm loss function
    # print(L)
    dw0 = (np.sum(y_-Y)*2)/len(Y) #Công thức tính đạo hàm của Loss function theo w0
    dw1 = (np.dot((y_-Y),X)*2)/len(Y) #Công thức tính đạo hàm của Loss function theo w1
    w0_ = w0_ - alpha*dw0
    w1_ = w1_ - alpha*dw1
    w0 = np.append(w0, w0_)
    w1 = np.append(w1, w1_)
    # print(w0_)
    # print(w1_)
    # print("\n")
    # diff = 1/2*sum((y_-Y)**2) - L[i-1]
    # if diff < 1e-3:
    #     iter_ = i
    #     break
end = dt.time()    
print('Thoi gian tinh toan:', end-start)
plt.plot(range(n), L)
plt.show()
```

```python
#Tính theo phương pháp gradient
w0_grd = w0[n-1]
w1_grd = w1[n-1]
print(w0_grd, w1_grd)
```

```python
#Tính theo công thức
w1_ct = (X.mean()*Y.mean()- (X*Y).mean())/(X.mean()*X.mean()-(X*X).mean())
w0_ct = Y.mean()-w1_ct*X.mean()
print(w0_ct, w1_ct)
```

```python
#Vẽ đồ thị theo phương pháp gradient
Y_ = w0[n-1] + w1[n-1]*X #Tính Y_ theo gradient
plt.scatter(dat3['x'], dat3['y'])
plt.plot(X, Y_)
plt.show()
```

```python
#Vẽ đồ thị theo công thức
Y_ = w0_ct + w1_ct*X #Tính Y_ theo công thức
plt.scatter(dat3['x'], dat3['y'])
plt.plot(X, Y_)
plt.show()
```

## DEMO5: Polynomial Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures as pf
from sklearn.linear_model import LinearRegression as lr
import seaborn as sns
print('Khai báo thành công')
```

```python
#Nhập dữ liệu
x = np.arange(0, 30)
y = [3, 4, 5, 7, 10, 8, 9, 10, 10, 23, 27, 44, 50, 63, 67, 60, 62, 70, 75, 88, 81, 87, 95, 100, 108, 135, 151, 160, 169, 179]
```

```python
x
```

```python
y = np.array(y) #Chuyển từ python List thành numpy array --> tăng tốc tính toán
y
```

```python
sns.scatterplot(x=x, y=y);
```

```python
#Chuyển đổi sang bậc cao (đa thức hóa biến x)
poly = pf(degree=2) # khai báo đa thức bậc 2
poly_X = poly.fit_transform(x.reshape(-1, 1)) #đa thức hóa
# poly_X
```

```python
#Tạo model regression
pr_model = lr() #Tạo đối tượng thuộc class LinearRegression
pr_model.fit(poly_X, y) #Fit (tính toán)
```

```python
#Dự đoán (prediction)
Y_pred = pr_model.predict(poly_X) 
# Y_pred
```

```python
#Tính r2score
import sklearn.metrics as metrics
metrics.r2_score(y,Y_pred)
```

```python
#Giả sử bậc 3
#Chuyển đổi sang bậc cao (đa thức hóa biến x)
poly = pf(degree=3) # khai báo đa thức bậc 2
poly_X = poly.fit_transform(x.reshape(-1, 1)) #đa thức hóa

#Tạo model regression
pr_model = lr() #Tạo đối tượng thuộc class LinearRegression
pr_model.fit(poly_X, y)

#Tạo giá trị y dự đoán
Y_pred = pr_model.predict(poly_X) 

#Vẽ hình
plt.scatter(x, y)
plt.plot(x, Y_pred, 'r')
plt.show()

print(metrics.r2_score(y,Y_pred))
```

```python
#Dự đoán với x bất kỳ
x1 = 30
X1 = np.array(x1) #Chuyển thành dạng numpy
polyX1 = poly.fit_transform(X1.reshape(-1, 1)) #Đa thức hóa
y1 = pr_model.predict(polyX1) #Quá trình dự đoán
print(y1)
```

# Sử dụng pipeline

```python
#Pipeline là 1 quy trình, tương tự như user function
from sklearn.pipeline import Pipeline

#Chuyển đổi sang bậc cao
poly = pf(degree=2)

#Tạo model regression
pr_model = lr()

#Tạo pipeline
pipeline = Pipeline(
        [
            ("polynomial_features", poly),
            ("linear_regression", pr_model)
        ]
    )

#Tính pipeline
pipeline.fit(x.reshape(-1,1), y)

#Tạo giá trị y dự đoán
Y_pred = pipeline.predict(x.reshape(-1, 1))

#Vẽ hình
sns.scatterplot(x=x, y=y)
sns.lineplot(x=x, y=Y_pred, color='r')
plt.show()
```

```python
x2 = 30
y2 = pipeline.predict(np.array(x2).reshape(-1, 1))
y2
```

## Demo 6: Regression multifeature

```python
area = [2100,1416,1534,852,1600,1985,1535,1050,2300,1200]
bedroom = [5,3,3,2,3,5,4,2,4,3]
floor = [1,2,2,1,2,1,2,1,2,2]
price = [460,232,315,178,329,420,330,195,450,250]
```

```python
Y = np.asarray(price) #Xử lý Y --> numpy array
#Xử lý X
X = np.append(np.array(area).reshape(-1, 1), np.array(bedroom).reshape(-1, 1), axis=1)
X = np.append(X, np.array(floor).reshape(-1, 1), axis=1)
X
```

```python
demo6_model = lr(normalize=True) #Tạo object hồi quy tuyến tính, có chuẩn hóa

demo6_model.fit(X, Y) #Tính toán model

# Dự đoán 
Y_ = demo6_model.predict(X)
# Y_

#Vẽ đồ thị
sns.scatterplot(area, price)
# temp = np.arange(10)*0
# x_ = np.append(np.array(area).reshape(-1, 1), temp.reshape(-1, 1), axis=1)
# x_ = np.append(x_, temp.reshape(-1, 1), axis=1)
sns.lineplot(X[:,0], Y_, color='r')
plt.show()
```

```python
#Sử dụng pipeline (kiểu mới)
from sklearn.preprocessing import StandardScaler
zscore = StandardScaler(with_mean=False)
demo6_model = lr()
#Tạo pipeline
pipeline = Pipeline(
        [
            ("StandardScaler", zscore),
            ("linear_regression", demo6_model)
        ]
    )
pipeline.fit(X,Y)
Y_ = pipeline.predict(X)

#Vẽ
sns.scatterplot(x=X[:,0], y=Y);
sns.lineplot(x=X[:,0], y=Y_, color='r');
```

```python
#Kiểm tra model có chính xác hay ko
from sklearn.metrics import r2_score

r2 = r2_score(Y, Y_)
r2
```

```python
dat71 = dat71.drop(columns = ['sex_male', 'smoker_no'])
dat71
```

```python
X = train.drop(columns=['charges'])
Y = train.loc[:,'charges']
```

```python
#Sử dụng pipeline (kiểu mới)
#Tạo chuẩn hóa
from sklearn.preprocessing import StandardScaler
zscore = StandardScaler(with_mean=False)

#Tạo model Linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()

#Tạo pipeline
from sklearn.pipeline import Pipeline
pipeline = Pipeline([("StandardScaler", zscore),
                     ("linear_regression", model)])

#Fit
pipeline.fit(X,Y)

#Predict
Y_ = pipeline.predict(test.drop(columns=['charges']))

#Tính r2score
import sklearn.metrics as metrics
metrics.r2_score(test.loc[:,'charges'], Y_)
```
