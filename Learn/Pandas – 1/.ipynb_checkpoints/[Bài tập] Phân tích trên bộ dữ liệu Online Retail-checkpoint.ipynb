# 2. Trích xuất dữ liệu các cột Description và Quantity lưu vào file OnlineRetail.csv

```python
df.loc[:, ['Description', 'Quantity']].to_csv('OnlineRetail.csv')
```

```python
pd.read_csv('OnlineRetail.csv').head()
```

# 3. Trích xuất dữ liệu 1000 dòng đầu tiên lưu vào file OnlineRetail.xlsx

```python
df.iloc[:1000].to_excel('OnlineRetail.xlsx', encoding = 'UTF-8')
```

```python
pd.read_excel('OnlineRetail.xlsx')
```

# 4. Trích xuất các bản ghi có số lượng từ 10 trở lên lưu vào file OnlineRetail.h5

```python
df['Quantity'] > 10
```

```python
df.loc[df['Quantity'] > 10].to_hdf('OnlineRetail.h5', 'table')
```

```python
pd.read_hdf('OnlineRetail.h5', 'table')
```

```python
pd.read_hdf('OnlineRetail.h5', 'table').loc[df['Quantity'] <= 10]
```

```python
check_quantity = pd.read_hdf('OnlineRetail.h5', 'table')['Quantity'] > 10
check_quantity.unique()
```

# 5. Trích xuất dữ liệu các phần tử từ dòng 1000 đến dòng 2000, các cột Quantity, InvoiceDate, UnitPrice lưu vào file OnlineRetail.json

```python
df.loc[1000:2000, ['Quantity', 'InvoiceDate', 'UnitPrice']].to_json('OnlineRetail.json')
```

```python
pd.read_json('OnlineRetail.json')
```

# 6. Trích xuất các bản ghi có số hóa đơn ‘536365’ lưu vào file OnlineRetail.html

```python
df.loc[df['InvoiceNo'] == '536365'].to_html('OnlineRetail.html')
```
