### 2. Xây dựng bảng Pivot table, với mỗi Số hóa đơn tính trung bình cộng số lượng các mặt hàng theo từng Quốc gia.

```python
import collections
```

```python
country_count_dataframe = pd.DataFrame.from_dict(dict(collections.Counter(df.Country)), orient = 'index', columns = ['Times of appearance']).sort_values(by = 'Times of appearance', ascending = False)
country_count_dataframe
```

### 3. Xây dựng bảng Pivot table, với mỗi Khách hàng cho biết số lượng mua hàng lớn nhất và nhỏ nhất theo Kho.

```python
df.head()
```

```python
df.dtypes
```

```python
quantity_of_customer_by_stock = df.pivot_table(values = 'Quantity', index = ['CustomerID', 'StockCode'], aggfunc = [np.min, np.max])
```

```python
quantity_of_customer_by_stock
```

### 4. Xây dựng bảng Pivot table, với mỗi Mã kho tính tổng số lượng các mặt hàng và trung bình cộng giá.

```python
sum_of_products_and_mean_of_price = df.pivot_table(values = ['Quantity', 'UnitPrice'], index = 'StockCode', aggfunc = {'Quantity': np.sum, 'UnitPrice': np.mean}) 
```

```python
sum_of_products_and_mean_of_price = sum_of_products_and_mean_of_price.rename(columns = { 'Quantity': 'Tổng số lượng mặt hàng', 'UnitPrice': 'Trung bình cộng giá'}, index = {'StockCode': 'Mã kho'})
```

```python
sum_of_products_and_mean_of_price
```
