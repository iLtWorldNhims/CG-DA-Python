# 1. Chuẩn bị dữ liệu

```python
# Đọc file dữ liệu 'FoodPrice_in_Turkey.csv'
import pandas as pd
df = pd.read_csv(r'C:\Users\MY PC\Documents\GitHub\CG-DA-Python\Datasets\FoodPrice_in_Turkey.csv')
df
```

```python
# Tách file 1 chứa 5000 bản ghi đầu tiên
df1 = df.loc[0:4999, :]
df1
```

```python
# Tách file 2 chứa các bản ghi còn lại
df2 = df.loc[5000:, :]
df2
```

```python
# Tách file 3 chứa thông tin giá với số dòng từ bản ghi 1000 đến 2000
df3 = df.loc[1000:2000, ['ProductId', 'Place', 'Month', 'Year', 'Price']]
df3
```

```python
df_test = df.iloc[1000:2000] # .loc include cả trái lẫn phải - còn .iloc include trái và exclude phải
df_test
```
