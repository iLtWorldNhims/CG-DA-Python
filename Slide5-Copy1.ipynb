# Biểu đồ tròn (bánh)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Nhập dữ liệu
data = pd.read_csv('GDPlist.csv', header = 0, encoding = 'ISO-8859-1')
data
```

```python
#Tổng quan về dữ liệu
data.info()
```

```python
data.loc[:,'Country'].unique()
```

```python
data.loc[:,'Continent'].unique()
```

```python
#Lọc ra các quốc gia South America
SA = data.loc[data.loc[:, 'Continent'] == 'South America']
SA
```

```python
#Nên dùng biểu đồ cột (bar)
plt.bar(SA.loc[:, 'Country'], SA.iloc[:, 2])
plt.show()
```

```python
SA = SA.sort_values(by = 'GDP (millions of US$)')
```

```python
plt.bar(x = SA.loc[:, 'Country'], height = SA.loc[:, 'GDP (millions of US$)'] / 1000)
plt.xticks(rotation = 45)
plt.ylabel('Tỷ USD')
plt.title('So sánh GDP các quốc gia Nam Mỹ')
plt.show()
```

Nhận xét:
*Trong 9 nước Nam Mỹ thì Brazil là nước có tổng sản phẩm quốc nội cao nhất và vượt lên hẳn các nước còn lại.*
*Ngoài ra, các nước còn lại có thể được chia thành 2 nhóm nhỏ hơn:*
+ Nhóm trung bình: Peru, Chile, Venezuela, Colombia và Argentina
+ Nhóm thấp: Paraguay, Uruguay và Ecuador

**2.Đánh giá tỉ lệ đóng góp GDP của Việt Nam trên tổng số GDP của 5 nước Đông Nam Á là Vietnam, Indonesia, Cambodia, Thailand và Malaysia.**

```python
data.loc[:, 'Continent'].unique()
```

```python
sea_ = ['\xa0Vietnam', '\xa0Indonesia', '\xa0Cambodia', '\xa0Thailand', '\xa0Malaysia']
sea_
```

```python
SEA = data.loc[data.loc[:, 'Country'].isin(sea_)]
SEA
```

```python
#Lọc ra 5 nước DNA
DNA_ = ['\xa0Vietnam', '\xa0Indonesia', '\xa0Cambodia', '\xa0Thailand', '\xa0Malaysia']

DNA = pd.DataFrame(columns=data.columns)
for country in DNA_:
    DNA = pd.concat([DNA, data.loc[data.loc[:,'Country']==country]]) #pd.concat([dt1, dt2], axis)
    # DNA = DNA.append(data.loc[data.loc[:,'Country']==country]) #deprecated 
DNA
```

```python
#Cách 2 nhanh hơn, dùng hàm query
DNA = data.query('Country in @DNA_')
DNA
```

```python
plt.pie(SEA.loc[:, 'GDP (millions of US$)'], labels = SEA.loc[:, 'Country'], autopct='%1.2f%%')
plt.title('So sánh GDP các nước DNA')
plt.show()
```
